Energy Industries:
specifically focuses on the energy-related industries involved in the production, extraction, and processing of energy resources. It includes sectors such as oil and gas extraction, coal mining, refining and processing of fossil fuels, and electricity generation. The "Energy Industries" category provides insights into the specific industries that contribute to energy production and supply.

Buildings: refers to the emissions associated with energy use in residential, commercial, and institutional buildings. It includes emissions from the consumption of electricity, natural gas, heating oil, and other fuels used for heating, cooling, lighting, and operating buildings. The "Buildings" category represents the energy-related emissions directly attributed to the building sector.

Manufacturing and Construction: includes emissions from industrial processes, manufacturing activities, and construction activities. It encompasses emissions generated during the production of goods and materials, as well as emissions resulting from construction projects. This category covers a wide range of industries, such as manufacturing, mining, chemical production, cement production, and other heavy industries.

```python
industry_dynamics=GHG.groupby('Industry').sum()
transp=industry_dynamics.transpose()
top_industries=["Energy Industries", "Road Transportation", "Agriculture", "Buildings and other Sectors", "Waste", "Land-use, land-use change and forestry"]
top_transposed=transp[top_industries]
top_transposed
```

```python
top_transposed.plot(kind='line',figsize=(9,7))
plt.xlabel('Year')
plt.ylabel('Emissions, million t.')
plt.title('Greenhouse Gas Emissions, Dynamics by Industry')
plt.legend(loc='upper right') 
plt.grid()
plt.show()
```
